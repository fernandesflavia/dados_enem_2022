```python
dados1 = dados.rename(columns={'NU_NOTA_CN': 'NOTA_CN','NU_NOTA_CH': 'NOTA_CH',
                                'NU_NOTA_LC': 'NOTA_LC','NU_NOTA_MT': 'NOTA_MT'})
```

```python
dados1 = dados.rename(columns={'TP_SEXO': 'SEXO','TP_COR_RACA':'RACA','TP_ESTADO_CIVIL':'ESTADO_CIVIL',
                             'TP_FAIXA_ETARIA':'IDADE','TP_NACIONALIDADE':'NACIONALIDADE','TP_ESCOLA':'ESCOLA' })
```

```python
dados1.head()
```

#Renomeando Registros (Linhas)

```python
dados1["RACA"] = dados1["RACA"].replace({0:"nao_declarado", 1:"branca", 2:"preta", 3:"parda", 4:"amarela", 5:"indigena"})
```

```python
dados1.head()
```

```python
dados1["ESTADO_CIVIL"] = dados1["ESTADO_CIVIL"].replace({0:"Nao informado", 1:"Solteiro(a)", 2:"Casado(a)", 3:"Divorciado(a)", 4:"Viúvo(a)"})
```

```python
dados1.head()
```

```python
dados1 = dados1.drop(columns=['TP_ANO_CONCLUIU','TP_ENSINO'],)
```

```python
dados1.head()
```

```python
dados1 = dados1.drop(columns=['TP_LOCALIZACAO_ESC',],)
```

```python
dados1.head()
```

```python
dados1["TP_LINGUA"] = dados1["TP_LINGUA"].replace({0:"Inglês", 1:"Espanhol"})
```

```python
dados1["ESCOLA"] = dados1["ESCOLA"].replace({1:"não_respondeu", 2:"pública", 3:"privada", 4:"exterior"})
```
